```python
b5=0
b4=0
b3=0
for i in range(len(by_year)):
    curdat=by_year.iloc[i]
    color = 'black'
    weight = int(curdat['MAG'])
    if curdat['MAG'] == 2:
        color = 'green'
    elif curdat['MAG'] == 3:
        color = 'yellow'
    elif curdat['MAG'] == 4:
        color = 'blue'
    elif curdat['MAG'] == 5:
        color = 'red'

    if curdat['ELAT'] == curdat['ELON'] == 0.0:
        tor=folium.Circle(location=(curdat['SLAT'],curdat['SLON']), radius=1, color=color,weight=weight)
        
    else:
        coords = [(curdat['SLAT'],curdat['SLON']),(curdat['ELAT'],curdat['ELON'])]
        tor = folium.PolyLine(locations=coords, weight=weight, color=color, tooltip=curdat.DATE)
    m.add_child(tor)

# print(f'F5 count: {b5}')
# print(f'F4 count: {b4}')
# print(f'F3 count: {b3}')
# print(f'Overall: {b5+b4+b3}')
```
