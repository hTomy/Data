```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df=pd.read_csv('C:/Users/tamas.hegedus/Desktop/Border_Crossing_Entry_Data.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df= df.drop(['Location','Port Code'], axis=1)
```

Bus Passengers, Buses, Pedestrians, Personal Vehicle Passengers, Personal Vehicles, Rail Containers Empty, Rail Containers Full, Train Passengers, Trains, Truck Containers Empty, Truck Containers Full, Trucks

```python
mean_of_travel = 'Personal Vehicles'
border = 'US-Mexico Border'
mot_filter = df['Measure'] == mean_of_travel
border_filter = df['Border'] == border
cropped_data = df[mot_filter & border_filter]

year = 2016
start_date = '01-01-' + str(year)
end_date = '31-12-' + str(year)
mask = (cropped_data['Date'] > start_date) & (cropped_data['Date'] <= end_date)

year_cropped = cropped_data.loc[mask]

ma=year_cropped.groupby(['Port Name'], as_index=False).max()

ba=ma.sort_values('Value',ascending=False).head(7)

plt.figure(figsize=(15,4))
plt.bar(ba['Port Name'], ba['Value'], color='green')
plt.xticks( rotation=20)
plt.title('Most crossings by ['+ mean_of_travel +'] in 2018 in ' + border)
```
