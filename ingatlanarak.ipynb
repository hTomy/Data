```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
import sys
```

```python
varos = 'szeged'
url = f'https://ingatlan.com/lista/kiado+lakas+{varos.lower()}?page=%s'
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.83 Safari/537.36'
}
```

```python
df = pd.DataFrame(columns=['price', 'location', 'dimensions', 'rooms', 'date', 'url'])
```

```python
resp = requests.get(url%1, headers = headers)
```

```python
i = 1
while True:
    resp = requests.get(url%i, headers = headers)
    i += 1
    soup = BeautifulSoup(resp.text, 'html.parser')
    if soup.find('div', {'class': 'no-results__title'}):
        sys.exit(0)
    lista = soup.find('div', {'class': 'resultspage__listings js-listings'})
    elements = lista.findAll('a', {'class': 'listing__link js-listing-active-area'})

    for current in elements:
        price_tag = current.find('div', {'class':'price'})
        price = int(''.join([a for a in price_tag.text if a.isdigit()]))

        address = current.find('div', {'class': 'listing__address'}).text


        dimensions_tag = current.find('div', {'class': 'listing__parameter listing__data--area-size'})
        dimensions = int(re.search(r'^[0-9]*',dimensions_tag.text.strip())[0])


        rooms = current.find('div', {'class': 'listing__parameter listing__data--room-count'}).text.strip()


        df = df.append(pd.Series({'price': price, 'location': address, 'dimensions': dimensions, 'rooms': rooms, 'date': pd.to_datetime('today').date(), 'url': current.get('href')}), ignore_index=True)
```
